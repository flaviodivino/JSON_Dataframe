Usando o Pandas para transformar um JSON em um Dataframe

```python
#importando as bilbiotecas
import json
import pandas as pd 
from pandas import DataFrame
from pandas.io.json import json_normalize
```

```python
# lemos o JSON em disco
arquivo = open("C://DatasetRestaurante//sample.txt","r").read()
conteudo = json.loads(arquivo)

# lendo o arquivo json com a opção orient = "records" para tentar ver quais as colunas que podem possuir arrays
# Esse é o cabeçado do Dataframe 'complemento', 'dets, 'emit', 'ide', 'infAdic','total','versaoDocumento'
# Observe que na coluna "dets" possui um colchete "[", isso significa que essa coluna possui um array
# Usei o head(2) para mostrar apena dois registros do dataframe
pd.read_json(arquivo,orient='records').head(2)
```

```python
#Normalizando o conteudo do JSON para ser transformado em um novo Dataframe
dataframe = json_normalize(conteudo, 'dets', ['complemento','emit', 'ide', 'infAdic','total','versaoDocumento'])
```

```python
#Observe que nesse novo Dataframe apareceram as colunas "nItem" e "prod" que eram um array dentro da coluna "dets"
dataframe.head(2)
```

```python
# Vamos transformar esse Dataframe novamente em JSON
novo_json = dataframe.to_json(orient='records')
```

```python
# Vamos ler novamente esse JSON
novo_arquivo = json.loads(novo_json)
```

```python
# E finalmente vamos normalizar esse JSON para deixar o Dataframe correto 
datafrane_final = json_normalize(novo_arquivo)
```
